```python
import numpy as np
import matplotlib.pyplot as plt

# Data extracted from the tables
methods = ["Enlarging PCC", "Reducing RD", "Regression", "One-hot Label", "Label Smoothing", 
           "Ours (Gaussian)", "Ours (Linear)", "Ours (Exp.)"]

pcc_values = [54.4, 15.1, 40.3, 37.6, 35.0, 53.7, 54.0, 56.3]  # PCC (%)
rd_values = [40.1, 29.9, 28.1, 43.3, 36.1, 24.0, 24.6, 25.6]  # RD (%)

# Highlighting "Ours" methods
highlight_colors_pcc = ['gray' if "Ours" not in method else 'blue' for method in methods]
highlight_colors_rd = ['gray' if "Ours" not in method else 'red' for method in methods]

# Create figure with two subplots side by side
plt.figure(figsize=(12,4))

# Horizontal bar plot for PCC values
plt.subplot(1, 2, 1)
bars_pcc = plt.barh(methods, pcc_values, color=highlight_colors_pcc)
plt.title('PCC Comparison')
plt.xlabel('PCC (%)')
plt.ylabel('Methods')
plt.xlim(0, 60)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adding data labels to the right of the bars
for bar, value in zip(bars_pcc, pcc_values):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{value:.1f}', 
             va='center', fontsize=10)

# Horizontal bar plot for RD values
plt.subplot(1, 2, 2)
bars_rd = plt.barh(methods, rd_values, color=highlight_colors_rd)
plt.title('RD Comparison')
plt.xlabel('RD (%)')
plt.ylabel('Methods')
plt.xlim(0, 50)
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Adding data labels to the right of the bars
for bar, value in zip(bars_rd, rd_values):
    plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{value:.1f}', 
             va='center', fontsize=10)

plt.tight_layout()
plt.show()
```
