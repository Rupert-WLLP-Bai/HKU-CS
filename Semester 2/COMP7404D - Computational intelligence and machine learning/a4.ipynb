### A3
The parity problem returns 1 if the number of inputs that are 1 is even, and 0 otherwise. Can a perceptron learn this problem for 3 inputs? 


*Your answer here*
### A3

No, a perceptron cannot learn the parity problem for 3 inputs. The parity problem requires determining if the number of 1s in the input is even (output 1) or odd (output 0).

The fundamental limitation is that a perceptron can only represent linearly separable functions, meaning it can only draw a single straight line (or hyperplane in higher dimensions) to separate the two classes.

For 3 inputs, the parity function maps:
- (0,0,0) → 1 (even number of 1s: 0)
- (0,0,1) → 0 (odd number of 1s: 1)
- (0,1,0) → 0 (odd number of 1s: 1)
- (0,1,1) → 1 (even number of 1s: 2)
- (1,0,0) → 0 (odd number of 1s: 1)
- (1,0,1) → 1 (even number of 1s: 2)
- (1,1,0) → 1 (even number of 1s: 2)
- (1,1,1) → 0 (odd number of 1s: 3)

These points form the vertices of a cube in 3D space, with alternating classifications. No single hyperplane can separate the points with even parity from those with odd parity.

To solve this problem, we would need a multi-layer network with at least one hidden layer, such as a multilayer perceptron (MLP) with nonlinear activation functions.

### A5
Suppose that the probability of five events are $P(first) = 0.5, P(second) = P(third) = P(fourth) = P(fifth) = 0.125$. Calculate the entropy and write down in words what this means. 


*Your answer here*
### A5

The entropy of a probability distribution is calculated as:
$H = -\sum_{i=1}^{n} P(i) \log_2 P(i)$

For the given probabilities:
- $P(first) = 0.5$
- $P(second) = P(third) = P(fourth) = P(fifth) = 0.125$

Calculating the entropy:
$H = -(0.5 \log_2 0.5 + 0.125 \log_2 0.125 + 0.125 \log_2 0.125 + 0.125 \log_2 0.125 + 0.125 \log_2 0.125)$
$H = -(0.5 \cdot (-1) + 0.125 \cdot (-3) + 0.125 \cdot (-3) + 0.125 \cdot (-3) + 0.125 \cdot (-3))$
$H = -((-0.5) + (-0.375) + (-0.375) + (-0.375) + (-0.375))$
$H = -(-2)$
$H = 2$ bits

This entropy value represents the average amount of information or uncertainty in this probability distribution. A value of 2 bits indicates moderate uncertainty - more than a uniform binary distribution (1 bit) but less than a uniform distribution across all five events (which would be about 2.32 bits). The first event has a higher probability (0.5) which reduces the overall entropy compared to if all events had equal probability.

For a logistic regression model with weights $w_0 = -6$, $w_1 = 0.05$, $w_2 = 1$:

1. **Probability of getting an A:**
    For a student with 40 hours of study time and 3.5 GPA:
    
    $z = w_0 + w_1 \times x_1 + w_2 \times x_2$
    $z = -6 + 0.05 \times 40 + 1 \times 3.5$
    $z = -6 + 2 + 3.5$
    $z = -0.5$
    
    $P(A) = \frac{1}{1 + e^{-z}} = \frac{1}{1 + e^{0.5}} \approx 0.3775$ or approximately 37.75%

2. **Hours needed for 50% chance of getting an A:**
    We need to find $x_1$ such that $P(A) = 0.5$

    $0 = -6 + 0.05 \times x_1 + 1 \times 3.5$

    $0 = -2.5 + 0.05x_1$

Suppose the features in your training set have very different scales. Which algorithms discussed in class might suffer from this, and how? What can you do about it?


*Your answer here*
### A10

When features have very different scales, several algorithms can be negatively affected:

1. **Distance-Based Algorithms:**
    - **K-Nearest Neighbors (KNN)**: Features with larger scales will dominate distance calculations
    - **K-Means**: Clusters will be biased toward features with larger scales

2. **Gradient-Based Methods:**
    - **Linear/Logistic Regression**: Convergence becomes slower and potentially unstable
    - **Neural Networks**: Gradient descent may oscillate or converge slowly due to different gradient magnitudes

3. **Regularization-Sensitive Methods:**
    - **Ridge and Lasso Regression**: Regularization will penalize features with smaller scales more heavily
    - **Support Vector Machines (SVM)**: Optimal hyperplane positioning will be biased toward features with larger scales

4. **Variance-Sensitive Methods:**
    - **Principal Component Analysis (PCA)**: Components will be dominated by high-variance features
    - **Linear Discriminant Analysis (LDA)**: Class separation will focus on high-variance features

Solutions include:

- **Standardization**: Transform features to have mean=0 and standard deviation=1
- **Min-Max Scaling**: Scale features to a specific range (e.g., [0,1])
- **Normalization**: Scale each sample to unit norm
- **Robust Scaling**: Use statistics that are robust to outliers (e.g., median, quantiles)
- **Log Transformation**: For skewed data with a wide range of values

Algorithms like Decision Trees and Random Forests are relatively resistant to different feature scales since they make decisions based on thresholds rather than distances or gradients.

What is the benefit of out-of-bag evaluation?

*Your answer here*
### A12

Out-of-bag evaluation provides several key benefits:

1. **Efficient use of data** - It allows estimation of model performance without requiring a separate validation set, maximizing the data used for training

2. **Unbiased error estimation** - Each observation is evaluated only using trees that did not have that observation in their bootstrap sample, providing an honest assessment similar to cross-validation

3. **Built-in cross-validation** - Automatically performs a form of cross-validation during the training process without additional computational cost

4. **Feature importance evaluation** - Enables calculation of feature importance by measuring how much error increases when a feature is permuted in the out-of-bag samples

5. **Model tuning** - Provides a reliable metric for hyperparameter optimization without requiring a separate validation split

6. **Computational efficiency** - Eliminates the need for explicit cross-validation procedures, saving computational resources while still giving reliable performance estimates

7. **Handling limited data** - Particularly valuable when working with small datasets where traditional train-test splits would significantly reduce training data

### A13

What is the difference between hard and soft voting classifiers?

*Your answer here*
### A13

Hard voting vs Soft voting classifiers:

**Hard Voting**:
- Implements a "majority rule" approach
- Each classifier gets one vote for its predicted class
- The class that receives the most votes becomes the final prediction
- Only uses the final class prediction from each classifier
- No consideration of prediction confidence or probability
- Like asking multiple experts to vote on a decision, with each expert having equal weight

**Soft Voting**:
- Uses predicted probabilities from each classifier
- Averages the probabilities for each class across all classifiers
- Chooses the class with the highest average probability
- Takes into account the confidence level of each classifier's predictions
- Requires classifiers that can output probability estimates
- Generally performs better than hard voting when classifiers are well-calibrated
- Like asking multiple experts to assign confidence scores to different options

Soft voting often outperforms hard voting because it retains and utilizes more information from the base classifiers, especially when classifiers have varying levels of confidence in their predictions for different instances.

````python
import numpy as np

class Perceptron(object):
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
        Learning rate (between 0.0 and 1.0)
    n_iter : int
        Passes over the training dataset.

    Attributes
    -----------
    w_ : 1d-array
        Weights after fitting.
    errors_ : list
        Number of misclassifications in every epoch.

    """
    def __init__(self, eta=0.01, n_iter=10):
        self.eta = eta
        self.n_iter = n_iter

    def fit(self, X, y):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
            Training vectors, where n_samples is the number of samples and
            n_features is the number of features.
        y : array-like, shape = [n_samples]
            Target values.

        Returns
        -------
        self : object

        """
        self.w_ = np.zeros(1 + X.shape[1])
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_[1:] += update * xi
                self.w_[0] += update
                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, -1)
````

As shown in function plot_decision_regions, the decision regions can be visualized by dense sampling via meshgrid. However, if the grid resolution is not enough, as artificially set below, the boundary will appear inaccurate.

Implement function plot_decision_boundary below to analytically compute and plot the decision boundary.

````python
def plot_decision_boundary(X, y, classifier):    
    
    # Calculate the decision boundary analytically
    x1_interval = np.linspace(X[:, 0].min() - 1, X[:, 0].max() + 1, 100)
    x2_interval = -(classifier.w_[1] * x1_interval + classifier.w_[0]) / classifier.w_[2]
    
    plt.plot(x1_interval, x2_interval, color='green', linewidth=4, label='boundary')
````
